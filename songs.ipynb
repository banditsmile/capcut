```python
import os
import json
files = os.listdir('songs')
files = [f for f in files if f.startswith('get_collection')]
data = []
for f in files:
    fo = open("songs/"+f,"r+")
    content = fo.read()
    fo.close()
    json_data = json.loads(content)
    print(len(json_data['data']['songs']))
    data= data+json_data['data']['songs']
    print(len(data))

open("total_songs.json","w+").write(json.dumps(data))
```
